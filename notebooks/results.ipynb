## Plotting the results for comparison

```python
import matplotlib.pyplot as plt
import numpy as np

eigen = [21.667, 44.964, 4.091]
el = [15, 19.003, 3.636]
fisher = [3.333, 4.361, 32.273]
```

|       |  EigenFace  | EigenFace(/top3) | FisherFace|
|-------|   ------    | --------         | ----- | 
|    Yale  |        21.667     |     15              |  3.33    |
| Yale B |     44.964  |      19.003    |   4.361    |
| CMU-PIE   |   4.091   | 3.636     |  32.273   |

```python
N = 3
ind = np.arange(N) 
width = 0.23

plt.figure(figsize=(7,5))
bar1 = plt.bar(ind, eigen, width, color = 'crimson')
bar2 = plt.bar(ind+width, el, width, color='dodgerblue')
bar3 = plt.bar(ind+width*2,fisher, width, color = 'pink')

plt.xlabel("Datasets")
plt.ylabel('Error Rates (in %)')
plt.title("Error Rates for Face Recognition")

plt.xticks(ind+width,['yale', 'yaleB', 'CMU-PIE'])
plt.legend( (bar1, bar2, bar3), ('EigenFaces', 'EigenFaces (except Top 3)', 'FisherFaces') )
plt.savefig("error_rates_plot.png")
plt.show()
```
